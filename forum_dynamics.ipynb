```python
from mailbox import mbox
import pandas as pd
import os, sys
import pytz
import time
```

### Import network attributes

```python
network = pd.read_csv('../data/network/vertex_attributes.csv')
cols = [u'name', u'degree',  u'in', u'out', u'core', u'core-in', u'core-out', u'closeness', u'betweenness', u'eigen_centrality', u'pagerank', u'neighborhood', u'avg_neighborhood', u'authority', u'hub', u'clustering_coef']
network = network[cols]
network.head()
```

### Top Question People

```python
labels =[u'core', u'core-in', u'core-out', u'closeness', u'betweenness', u'eigen_centrality', u'pagerank', u'neighborhood', u'authority', u'hub', u'clustering_coef']

questioners = network.drop(labels, axis=1)
top_questioners = questioners.sort_values(by=['in', 'out', 'avg_neighborhood'], ascending=[0, 0, 1])
top_questioners.head(20)
```

### Top Answer People

```python
labels = [u'degree',  u'in', u'core', u'core-in', u'core-out', u'closeness', u'betweenness', u'eigen_centrality', u'pagerank', u'neighborhood', u'authority', u'hub']

answerers = network.drop(labels, axis=1)
top_answerers = answerers.sort_values(by=['out', 'clustering_coef', 'avg_neighborhood'], ascending=[0, 1, 1])
top_answerers.head(20)
```

### Top Posters

```python
posters = pd.read_csv('../data/network/posting_statistics.csv')
top_posters = posters.sort_values(by=['posts'], ascending=[0])
top_posters.head(10)
```

### Top Discussion Starters

### Top Repliers

```python
posters = pd.read_csv('../data/network/posting_statistics.csv')
top_repliers = posters.sort_values(by=['replies'], ascending=[0])
top_repliers.head(10)
```

### Top Experts

```python
labels = [u'degree',  u'in', u'out', u'core', u'core-in', u'core-out', u'neighborhood', u'avg_neighborhood']

centralities = network.drop(labels, axis=1)
top_experts = centralities.sort_values(by=['pagerank'], ascending=[0])
top_experts.head(15)
```

### Top hubs

```python
top_hubs = centralities.sort_values(by=['hub'], ascending=[0])
top_hubs.head(15)
```

### Importance relative to other important people

```python
top_import = centralities.sort_values(by=['eigen_centrality'], ascending=[0])
top_import.head(15)
```

### Top bridges

```python
top_bridges = centralities.sort_values(by=['betweenness'], ascending=[0])
top_bridges.head(15)
```

### Most "clique-ish"

```python
cliques = centralities[centralities['clustering_coef'] < 1]
top_cliques = cliques.sort_values(by=['clustering_coef'], ascending=[0])
top_cliques.head(15)
```

### Least "clique-ish"

```python
cliques = centralities[centralities['clustering_coef'] > 0]
least_cliques = cliques.sort_values(by=['clustering_coef'], ascending=[1])
least_cliques.head(15)
```
